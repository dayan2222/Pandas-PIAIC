```
import pandas as pd
```

```
tables = pd.read_html('https://www.psx.com.pk/psx/announcement/financial-announcements')
# tables variable type is not dataframe, its list
print(type(tables))
print(len(tables))
```

```
stock = tables[0]
stock
```
